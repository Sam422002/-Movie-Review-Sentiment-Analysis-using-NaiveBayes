```python
# first remove the html tags
import re  # this is the regex library
clean = re.compile('<.*?>')
re.sub(clean , ' ',df.iloc[2].review)
```

```python
def clean_text(text):
    clean = re.compile('<.*?>')
    return re.sub(clean , '',text)
    
```

```python
df['review'] = df['review'].apply(clean_text)
```

```python
def lower_val(text):
    return text.lower()
```

```python
df['review'] = df['review'].apply(lower_val)
```

```python
df
```

```python
def remove_special(text):
    x =" "
    for i  in text:
        if i.isalnum():
            x = x +i
        else:
            x = x+ " "
    return x
```

```python
df['review'] = df['review'].apply(remove_special)
```

```python
df
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stop = stopwords.words("english")
```

```python
def remove_stopwords(text):
    x = []
    for i in text.split():
        if i not in stop:
            x.append(i)
    y = x[:]
    x.clear()
    return y
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
df
```

```python
from nltk.stem.porter import PorterStemmer
pr = PorterStemmer()
```

```python
y = []
def stem_words(text):
    
    for i in text:
        y.append(pr.stem(i))
    z = y[:]
    y.clear()
    return z
```

```python
df['review'] = df['review'].apply(stem_words)
```

```python
# steming means taking the words to its original form
```

```python
df
```

```python
def join_back(text):
    return " ".join(text)
```

```python
df['review'] = df['review'].apply(join_back)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from tqdm import tqdm

# --- Step 1: Initialize models ---
models = {
    
    "MultinomialNB": MultinomialNB(),
    "BernoulliNB": BernoulliNB()
}

# --- Step 2: Train all models with progress bar ---
print("Training models...\n")
for name in tqdm(models, desc="Training Progress", unit="model"):
    model = models[name]
    model.fit(x_train, y_train)

print("\n✅ All models trained successfully!")

# --- Step 3: Predict with progress bar ---
predictions = {}

print("\nMaking predictions...\n")
for name, model in tqdm(models.items(), desc="Prediction Progress", unit="model"):
    predictions[name] = model.predict(x_test)

# --- Step 4: Store predictions in separate variables if needed ---

y_pred_Ml = predictions["MultinomialNB"]
y_pred_BR = predictions["BernoulliNB"]

print("\n✅ All predictions completed successfully!")
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import accuracy_score

# Example for one model
accuracy = accuracy_score(y_test, y_pred_GS)
print("Accuracy (GaussianNB):", accuracy)
```

```python
# Example for one model
accuracy_ml = accuracy_score(y_test, y_pred_Ml)
print("Accuracy (GaussianNB):", accuracy_ml)
```

```python
# Example for one model
accuracy_BR = accuracy_score(y_test, y_pred_BR )
print("Accuracy (GaussianNB):", accuracy_BR)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

for name, y_pred in predictions.items():
    acc = accuracy_score(y_test, y_pred)
    pre = precision_score(y_test, y_pred, average='weighted')
    rec = recall_score(y_test, y_pred, average='weighted')
    f1  = f1_score(y_test, y_pred, average='weighted')
    print(f"\n{name}: Accuracy={acc:.4f}, Precision={pre:.4f}, Recall={rec:.4f}, F1={f1:.4f}")
```
